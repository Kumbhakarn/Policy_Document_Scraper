```python
# Define your existing folder path
folder = r"E:\Project\hdfc_policy_scraper\HDFC_Policy_Documents\html_data"
os.makedirs(folder, exist_ok=True)  # Ensure it exists

driver = webdriver.Chrome()
driver.get("https://www.hdfclife.com/policy-documents")
driver.maximize_window()
time.sleep(5)

# Locate all policy sections
elems = driver.find_elements(By.CLASS_NAME, "accordMenuContWrp")
print(f"{len(elems)} items found")

for index, elem in enumerate(elems):
    try:
        # Extract heading text
        heading = elem.find_element(By.TAG_NAME, "h2").text.strip()
    except:
        heading = f"document_{index}"

    # Clean and format filename
    filename = re.sub(r'[^\w\s-]', '', heading)  # Remove special characters
    filename = re.sub(r'\s+', '_', filename)     # Replace spaces with underscores
    filepath = os.path.join(folder, f"{filename}.html")

    # Get and save outer HTML
    d = elem.get_attribute('outerHTML')
    with open(filepath, 'w', encoding="utf-8") as f:
        f.write(d)

driver.quit()
```

```python
# Folder paths
html_folder = r"E:\Project\hdfc_policy_scraper\HDFC_Policy_Documents\html_data"
pdf_save_folder = r"E:\Project\hdfc_policy_scraper\HDFC_Policy_Documents"
os.makedirs(pdf_save_folder, exist_ok=True)

uin_pattern = re.compile(r'101N\d{3}V\d{2}')
global_counter = 1

# List HTML files
html_files = [f for f in os.listdir(html_folder) if f.endswith('.html')]
print(f"Found {len(html_files)} HTML files")

for html_file in html_files:
    with open(os.path.join(html_folder, html_file), 'r', encoding='utf-8') as f:
        soup = BeautifulSoup(f, 'html.parser')
    
    pdf_links = []
    for a in soup.find_all('a', href=True):
        href = a['href']
        if href.lower().endswith('.pdf'):
            if href.startswith('/'):
                href = 'https://www.hdfclife.com' + href
            pdf_links.append(href)

    print(f"\n{html_file}: {len(pdf_links)} PDF links found")

    saved_count = 0
    for link in pdf_links:
        if saved_count >= 5:
            break

        try:
            response = requests.get(link, timeout=20)
            response.raise_for_status()

            # Try to find UIN from URL first
            uin_match = uin_pattern.search(link)
            uin = None

            if not uin_match:
                # Try reading PDF content
                pdf_reader = PdfReader(BytesIO(response.content))
                text = ''
                for page in pdf_reader.pages:
                    text += page.extract_text() or ''
                    if len(text) > 2000:  # Don't read full PDF
                        break
                uin_match = uin_pattern.search(text)

            if uin_match:
                uin = uin_match.group()
                filename = f"{uin}_{global_counter}.pdf"
                filepath = os.path.join(pdf_save_folder, filename)

                with open(filepath, 'wb') as f:
                    f.write(response.content)

                print(f"  ✔ Saved: {filename}")
                global_counter += 1
                saved_count += 1
            else:
                print(f"  ✖ Skipped (No UIN): {link}")

        except Exception as e:
            print(f"  ✖ Failed to process: {link} ({e})")

print(f"\n✅ Done! Total PDFs saved with UINs: {global_counter - 1}")
```

```python
# --- CONFIGURATION ---
html_folder     = r"E:\Project\hdfc_policy_scraper\HDFC_Policy_Documents\html_data"
save_dir        = r"E:\Project\hdfc_policy_scraper\HDFC_Policy_Documents"
os.makedirs(save_dir, exist_ok=True)

# Regex for UIN (e.g. 101N134V25 or 101Y122V05)
uin_pattern = re.compile(r"\d{3}[NY]\d{3}V\d{2}")

pdf_counter = 1

# --- PROCESS EACH HTML ---
html_files = [f for f in os.listdir(html_folder) if f.lower().endswith('.html')]

for html_file in html_files:
    path = os.path.join(html_folder, html_file)
    with open(path, 'r', encoding='utf-8') as f:
        soup = BeautifulSoup(f, 'html.parser')

    pdf_links = []
    for a in soup.find_all('a', href=True):
        href = a['href'].strip()
        if href.lower().endswith('.pdf'):
            if href.startswith('/'):
                href = 'https://www.hdfclife.com' + href
            pdf_links.append(href)

    print(f"\n{html_file}: {len(pdf_links)} PDF links found")
    saved_count = 0

    for pdf_url in pdf_links:
        if saved_count >= 5:
            break
        try:
            pdf_response = requests.get(pdf_url, timeout=20)
            pdf_response.raise_for_status()

            # Try extracting UIN from PDF text
            try:
                reader = PdfReader(BytesIO(pdf_response.content))
                text = ""
                for page in reader.pages:
                    text += page.extract_text() or ""
                uin_match = re.search(r"\d{3}[NY]\d{3}V\d{2}", text)
            except Exception:
                uin_match = None

            # If not found in content, try extracting UIN from URL filename
            if not uin_match:
                uin_match = re.search(r"\d{3}[NY]\d{3}V\d{2}", pdf_url)

            if uin_match:
                uin = uin_match.group(0)
                filename = f"{uin}_{pdf_counter}.pdf"
                pdf_path = os.path.join(save_dir, filename)
                with open(pdf_path, "wb") as f_out:
                    f_out.write(pdf_response.content)
                print(f"✔ Saved: {filename}")
                pdf_counter += 1
                saved_count += 1
            else:
                print(f"✖ Skipped (No UIN): {pdf_url}")

        except Exception as e:
            print(f"✖ Failed to process: {pdf_url}  ({e})")

print(f"\nDone! Total PDFs saved with UINs: {pdf_counter - 1}")
```

```python
# --- CONFIG ---
html_folder = r"E:\Project\hdfc_policy_scraper\HDFC_Policy_Documents\html_data"
output_csv  = r"E:\Project\hdfc_policy_scraper\HDFC_Policy_Documents\policies.csv"

# List your 14 section names in the same order as the HTML files are named/sorted
section_names = [
    "Savings & Investment (Live)",
    "Retirement (Live)",
    "Child Education (Live)",
    "Protection (Live)",
    "Health (Live)",
    "ULIP (Live)",
    "Pension (Live)",
    "Term (Live)",
    "Traditional Savings (Live)",
    "Group (Live)",
    "Annuity (Live)",
    "Child Plans (Live)",
    "Micro Insurance (Live)",
    "Others (Live)"
]

# Regex to catch UINs like 101N134V25 or 101Y122V05
uin_regex = re.compile(r"\b\d{3}[NY]\d{3}V\d{2}\b", re.IGNORECASE)

records = []

# Get sorted list of HTML files (so index matches section_names)
html_files = sorted([f for f in os.listdir(html_folder) if f.lower().endswith(".html")])

for idx, html_file in enumerate(html_files):
    section = section_names[idx] if idx < len(section_names) else "Unknown"
    path = os.path.join(html_folder, html_file)
    with open(path, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "html.parser")

    # Find all policy links
    for a in soup.find_all("a", href=True):
        href = a["href"].strip()
        if not href.lower().endswith(".pdf"):
            continue

        name = a.get_text(strip=True) or os.path.basename(href)
        # Try UIN in the link text first, then the URL
        uin_match = uin_regex.search(name) or uin_regex.search(href)
        uin = uin_match.group(0) if uin_match else ""

        records.append({
            "Policy Name": name,
            "UIN": uin,
            "Policy Type": section,
            "PDF URL": href
        })

# Create DataFrame and save to CSV
df = pd.DataFrame(records, columns=["Policy Name","UIN","Policy Type","PDF URL"])
df.to_csv(output_csv, index=False, encoding="utf-8-sig")

print(f"Extracted {len(df)} policies; written to:\n  {output_csv}")
```
